##### Next word prediction is a natural language processing (NLP) task that involves predicting the most likely word to follow a given sequence of words in a sentence or text. It's a subtask of language modeling, where the goal is to understand and generate human-like language patterns. This task is essential in various applications, including Virtual Keyboards,Text Completion,Grammar and Style Correction.

```python
text="""Mohandas Karamchand Gandhi was born on October 2, 1869.
He was an Indian lawyer and political ethicist.Gandhi played a key role in India's independence from British rule.
Known as Mahātmā, meaning 'great-souled,' he employed nonviolent resistance and inspired civil rights movements globally.
Raised in Gujarat, India, Gandhi studied law in London. Later, he lived in South Africa for 21 years, where he first used nonviolent resistance for civil rights.
Returning to India in 1915, Gandhi became a leader in the Indian National Congress. He advocated for self-rule, easing poverty, women's rights, and religious amity.
He led campaigns, including the Dandi Salt March in 1930, challenging British policies. Despite imprisonment, he continued to champion India's independence.
However, his vision of a unified, religiously pluralistic India faced challenges in the 1940s with the demand for a separate Muslim state.
After India gained independence in 1947, communal violence erupted.
Gandhi, troubled by the strife, undertook hunger strikes for peace. He was assassinated on January 30, 1948, by Nathuram Godse, a Hindu nationalist.
Gandhi's birthday is celebrated as Gandhi Jayanti, and he is revered as the Father of the Nation in post-colonial India.
His life symbolizes nonviolence and political activism, leaving a lasting impact on the world.
"""
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from nltk.tokenize import sent_tokenize
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import warnings
warnings.filterwarnings('ignore')
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index # each Unique word is converted into tokens
```

```python
len(tokenizer.word_index)   #Total unique words in vocablury
```

```python
for  sentence in sent_tokenize(text):
    print(sentence)
```

```python
for  sentence in text.split('\n'):
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequences = []
for sentence in sent_tokenize(text):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenized_sentence)):
         input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
X = padded_input_sequences[:,:-1]
X.shape
```

```python
y = padded_input_sequences[:,-1]
y.shape
```

### This is a classification problem so i need  to convert these 193 numbers into classes

```python
#Doing one hot encoding of all the classes sinse we have 139 unique words i am making 140 classes cuz it strarts with zero class
y = to_categorical(y,num_classes=140)
y.shape
```

```python
model = Sequential()
model.add(Embedding(140, 150, input_length=21))
model.add(LSTM(150))
model.add(Dense(140, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(X,y,epochs=80)
```

```python
def pred(text):


  for i in range(5):
    # tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    # padding
    padded_token_text = pad_sequences([token_text], maxlen=21, padding='pre')
    # predict
    pos = np.argmax(model.predict(padded_token_text))

    for word,index in tokenizer.word_index.items():
      if index == pos:
        text = text + " " + word
        print(text)
```
